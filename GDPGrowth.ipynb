```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv(r"C:\Users\Divyansh Sirohi\Desktop\CRYPTONITE\India_GDP_Data.csv")
X = data[['Year']]
y = data['Percentage_Growth']
```

```python
model = LinearRegression()

model.fit(X, y)
```

```python
input_year = int(input("Enter a year for GDP growth prediction: "))

input_data = pd.DataFrame({'Year': [input_year]})

predicted_gdp_growth = model.predict(input_data)

print(f'Predicted GDP Growth for {input_year}: {predicted_gdp_growth[0]}')
```

```python
plt.scatter(X, y, color='black', label='Actual Data')
plt.scatter(input_year, predicted_gdp_growth, color='red', marker='x', label='Predicted Data')
plt.plot(X, model.predict(X), color='blue', label='Linear Regression Line')
plt.xlabel('Year')
plt.ylabel('Percentage Growth')
plt.title('Linear Regression on India\'s GDP Growth')
plt.legend()
plt.show()
```
